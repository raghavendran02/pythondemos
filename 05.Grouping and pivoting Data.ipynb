```python
emp.groupby('department_id').agg({'salary':'sum','employee_id':'size'})
```

```python
emp.groupby('department_id').agg({'salary':'sum','employee_id':'size'}).\
rename(columns={'salary':'sum_of_salary','employee_id':'count_of_employees'})
```

```python
emp.groupby('department_id').agg({'salary':['sum','mean'],'employee_id':'size'})
```

```python
emp.groupby(['department_id','job_id']).agg({'salary':'sum'}).\
rename(columns={'salary':'sum_of_salary'})
```

```python
#Ouputs an aggregate in long format 
#Values Argument : column on which aggregate to be applied
# aggfunc Arg : which aggregate to be applied
# index : grouping column(s) to be displayed accros the rows
# columns : grouping column(s) to be displayed accross the columns
emp.pivot_table(values = 'salary',index = 'department_id',columns='manager_id',aggfunc='sum')
```

```python
emp.pivot_table(values = 'salary',index = 'department_id',columns='job_id',aggfunc='sum')
```
