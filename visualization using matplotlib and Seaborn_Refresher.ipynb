## Visualizing Data using Python

Visualization is important when we are working with data. A large dataset cam be compressed into one single viusal depicting more information. We are going to use matplotlib library for Visualization.


 - Bar charts : Continuous Numeric vs categorical
 - Line charts : discrete Numeric
 - Histograms : Freqeuncy distribution
 - Box plots :  Distribution of data

```python
# import matplotlib's pyplot 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
import seaborn as sns
```

### Line chart
Line chart - is a type of chart used to visualize the value of something over time. For example, a finance department may plot the change in the amount of cash the company has on hand over time.

I want to plot the population of india and pakistan from 1960 to 2010 at the gap of 10 years.

```python
# creating a array of year

year = np.array([1960, 1970, 1980, 1990, 2000, 2010])

# creating a array of pakistan' population

pop_pakistan = np.array([44.91, 58.09, 78.07, 107.7, 138.5, 170.6])

# creating a array of india's population
pop_india = np.array([449.48, 553.57, 696.783, 870.133, 1000.4, 1309.1])
```

```python
# Line 1 - Presenting pakistan's population
plt.plot(year, pop_pakistan, color='green');
```

```python
# Line 2 - Presenting india's population
plt.plot(year, pop_india, color='orange');
```

```python
# Line 1 - Presenting pakistan's population
plt.figure(figsize=(10, 5))
plt.plot(year, pop_pakistan, color='green', label = 'population of pak')

# Line 2 - Presenting india's population
plt.plot(year, pop_india, color='orange', label = 'population of india')

plt.xlabel('Year')
plt.ylabel('Population in million')
plt.title('Pakistan - India Population till 2010')
plt.legend();
```

### Bar graph - Between Numeric and Categorical column

Bar Chart: it is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent. The bars can be plotted vertically or horizontally. It is used for Comparison among the item.

```python
df=pd.read_csv("diamonds.csv")
```

```python
df.shape
```

```python
df.head()
```

I would like to plot average cost for each diamond cut type . Here cut is categorical column and price is numerical column

```python
df_cut = df.groupby(['cut'])['price'].mean().reset_index()

# I can round off the price value to 2 decimal value. I will also arrange the data in descending order of price

df_cut['price'] = round(df_cut['price'], 2)

df_cut = df_cut.sort_values(by = 'price', ascending = False)

df_cut
```

```python
plt.bar(df_cut.cut,df_cut.price, width = 0.4)
plt.ylabel('Average price')
plt.xlabel('Type of cut')
plt.title('Cutwise Average diamond price');
```

```python
plt.barh(df_cut.cut,df_cut.price, height = 0.4)
plt.ylabel('Average price')
plt.xlabel('Type of cut')
plt.title('Cutwise Average diamond price');
```

```python
# Plotting the Bar chart
plt.figure(figsize=(10, 5))
ax = sns.barplot(x="cut", y="price", data= df_cut)

# Giving Y axis lable and X axis lable and chart title

ax.set(xlabel='Type of Cut', ylabel='Average price');
```

```python
# Plotting the Bar chart
plt.figure(figsize=(10, 5))
ax = sns.barplot(x="price", y="cut", data= df_cut)

# Giving Y axis lable and X axis lable and chart title

ax.set(xlabel='Average price', ylabel='Type of Cut');
```

```python
#Stacked and Grouped bar graph
```

```python
df_cut_color = df.groupby(['cut','color'])['price'].mean().reset_index()

# I can round off the price value to 2 decimal value. I will also arrange the data in descending order of price

df_cut_color['price'] = round(df_cut_color['price'], 2)

df_cut_color = df_cut_color.sort_values(by = 'price', ascending = False)

df_cut_color
```

```python
df_cut_color_final = df_cut_color.pivot("cut", "color", "price")

df_cut_color_final.head()
```

```python
#Stacked bar graph
df_cut_color_final.plot(kind= "bar", stacked = True, figsize = (15,5))
ax = plt.subplot()

ax.legend(loc='upper center', ncol=4);
```

```python
#Grouped bar graph
df_cut_color_final.plot(kind= "bar", stacked = False, figsize = (15,5))
ax = plt.subplot()

ax.legend(loc='upper center', ncol=4);
```

```python
# Plotting the grouped Bar chart using seaborn
plt.figure(figsize=(15, 5))
ax = sns.barplot(x="cut", y="price", hue = 'color', data= df_cut_color)

# Giving Y axis lable and X axis lable and chart title

ax.set(xlabel='Type of Cut', ylabel='Average price',title = 'Average Price per color and cut');
```
