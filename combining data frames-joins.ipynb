Q1 Find out names of employees and their names of the department in which they are working

```python
pd.merge(emp, dept, left_on="department_id", right_on="dept_id",how="inner")\
[["last_name","department_name"]]
```

Q2 Find out Names of departments and the Names of their managers who are managing those departments

```python
pd.merge(dept, emp, right_on="employee_id", left_on="manager_id")\
[["department_name","last_name"]].rename(columns = {"last_name":"HOD"})
```

Q3.For every department(Name), fetch the average salary in each department 

```python
df1 = emp[["department_id","salary"]].groupby("department_id").\
mean().reset_index().rename(columns={"salary":"dept_avg_salary"})
df1
```

```python
pd.merge(df1,dept, left_on = "department_id",right_on = "dept_id")\
[["department_name","dept_avg_salary"]]
```

Q4 Find out names of employees and their names of the department in which they are working. Also get the include those employees who are not working

```python
pd.merge(emp, dept, left_on="department_id", right_on="dept_id", how="left")\
[["last_name","department_name"]]
```

Q4 Find out names of employees and their names of the department in which they are working. Also get the include those departments which do not have any employees

```python
pd.merge(emp, dept, left_on="department_id", right_on="dept_id", how="right")\
[["last_name","department_name"]]
```
