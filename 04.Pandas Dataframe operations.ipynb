```python
# filtering rows
#Names and ID of all students who have 
#scored less than 70 in course 1 and above 70 in  Course2
# Condition : (gd.Marks_Course1 <70) &(gd.Marks_Course2 >70)
#['Name','StudentID','Marks_Course1','Marks_Course2']
```

```python
gd[(gd.Marks_Course1<70) & (gd.Marks_Course2>70)]\
[['Name','StudentID','Marks_Course1','Marks_Course2']]
```

```python
#Obtain the Names of the students who have taken the exam in MAR, APR
```

```python
gd[(gd.ExamMonth==3) | (gd.ExamMonth==4)]['Name']
```

```python
#NAmes of the students who have attended exam in months other than march
gd[~(gd.ExamMonth==3)]['Name']
```

```python
# using query function
gd.query("(Marks_Course1 <70) &(Marks_Course2 >70)")\
[['Name','StudentID','Marks_Course1','Marks_Course2']]
```

```python
gd[(gd.Marks_Course2 > 80) | (gd.Marks_Course3 > 80)][['Name']]
```

```python
#Student who has scored above 80 in second course 
# and scored below 80 in the third
gd.query("(Marks_Course2 >80) &(Marks_Course3 <80)")\
[['Name','StudentID','Marks_Course2','Marks_Course3']]
```
